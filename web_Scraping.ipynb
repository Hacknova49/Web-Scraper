# MULTIPLE SEASONS 

```python
#loop through range of seasons and insert new column 
#creating list of leagues
leagues = ['E0','E2','E3']
frames=[]
for leagues in leagues:
    for season in range(15,25):
        df=pd.read_csv(root+str(season)+str(season+1)+'/'+leagues+'.'+'csv')
        df.insert(1,'season',season)
        frames.append(df)
        
```

# ORGANIZE ALL THE DATA IN A DICTIONARY 

```python
# create a dictionary with original name of league as key
dict_countries = {
              'Spanish La Liga':'SP1', 'Spanish Segunda Division':'SP2',
              'German Bundesliga':'D1', 
              'English Premier League':'E0', 'English League 1':'E2', 'English League 2':'E3',
             }
```

```python
dict_countries['Spanish La Liga']
```

```python
for leagues in dict_countries:
    print(dict_countries[leagues])
```

```python
for leagues in dict_countries:
    print(dict_countries[leagues])
```

```python
dict_historical_data={}

#leagues = ['E0','E2','E3']
for leagues in dict_countries:
    frames=[]
    for season in range(15,25):
        df=pd.read_csv(root+str(season)+str(season+1)+'/'+dict_countries[leagues]+'.'+'csv')
        df.insert(1,'season',season)
        frames.append(df)
    df_concat=pd.concat(frames)
    dict_historical_data[leagues]=df_concat
        
```
